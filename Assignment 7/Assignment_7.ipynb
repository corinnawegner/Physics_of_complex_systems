```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

```python
def flow(y, t, sigma, r, b):
    dx = sigma * ( y[1] - y[0] )
    dy = r*y[0] - y[1] - y[0]*y[2]    
    dz = y[0]*y[1] - b*y[2]
    return [dx, dy, dz]
```

```python
sigma = 10
b = 8/3 
list_r = [0.5, 2, 20, 24, 28]
list_y = []
t = np.linspace(0,100,4000)
y0 = (5, 5, 10)

def plot_trajectories(dim):
    plt.figure(figsize=(30,8))
    for r in list_r:
        y = odeint(flow, y0, t, args=(sigma, r , b))
        #list_y.append(y)
        plt.plot(t, y[:,dim], label = f"r: {r}")
    plt.xlabel("time")
    plt.legend()

plot_trajectories(1)
plt.ylabel("y")
```

```python
plot_trajectories(0)
plt.ylabel("x")
```

```python
def plot_3D():
    fig = plt.figure(figsize=(10,14))
    ax = fig.add_subplot(projection='3d')
    for r in list_r:
        y = odeint(flow, y0, t, args=(sigma, r , b))
        ax.plot(y[:,0], y[:,1], y[:,2], label=f'r: {r}')  
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title('Trajectories ')
    ax.legend()
    plt.tight_layout()
plot_3D()
```
